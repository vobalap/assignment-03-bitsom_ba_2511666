```
# ============================================================
#   TASK 1 — File Read & Write Basics
#   Theme: Product Explorer & Error-Resilient Logger
# ============================================================

# ─────────────────────────────────────────
# PART A — WRITE & APPEND
# ─────────────────────────────────────────

# Step 1: Write 5 student notes to the file using write mode ('w')
notes = [
    "Topic 1: Variables store data. Python is dynamically typed.",
    "Topic 2: Lists are ordered and mutable.",
    "Topic 3: Dictionaries store key-value pairs.",
    "Topic 4: Loops automate repetitive tasks.",
    "Topic 5: Exception handling prevents crashes."]

with open("python_notes.txt", "w", encoding="utf-8") as file:
    file.writelines(notes)

print("File written successfully.")

# Step 2: Append 2 more lines using append mode ('a')
extra_notes = ["Topic 6: Functions help us reuse code and keep things organized.",
    "Topic 7: Modules let us import and use pre-built Python tools."]

with open("python_notes.txt", "a", encoding="utf-8") as file:
    file.writelines(extra_notes)

print("Lines appended.")


# ─────────────────────────────────────────
# PART B — READ, COUNT & SEARCH
# ─────────────────────────────────────────

# Step 3: Read the file and print each line with a line number
print("--- File Contents ---")

with open("python_notes.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

for i, line in enumerate(lines, start=1):
    print(f"{i}. {line.strip()}")

# Step 4: Count and print total number of lines
print(f"Total number of lines: {len(lines)}")

# Step 5: Ask user for a keyword and search (case-insensitive)
keyword = input("Enter a keyword to search: ")

print(f"--- Lines containing '{keyword}' ---")

matches = [line.strip() for line in lines if keyword.lower() in line.lower()]

if matches:
    for match in matches:
        print(f"  → {match}")
else:
    print(f"  No lines found containing the keyword '{keyword}'. Try a different word!")
```

```
# ============================================================
#   TASK 2 — API Integration
#   Theme: Product Explorer & Error-Resilient Logger
#   API: https://dummyjson.com/products
# ============================================================

# Install requests if not already available
# !pip install requests

import requests

# ─────────────────────────────────────────
# STEP 1 — Fetch and Display 20 Products
# ─────────────────────────────────────────

print("=" * 70)
print("STEP 1 — Fetching 20 Products from DummyJSON API")
print("=" * 70)

url = "https://dummyjson.com/products?limit=20"
response = requests.get(url)
data = response.json()

products = data["products"]

# Print formatted table header
print(f"{'ID':<4} | {'Title':<30} | {'Category':<14} | {'Price':<9} | {'Rating'}")
print(f"{'-'*4}-+-{'-'*30}-+-{'-'*14}-+-{'-'*9}-+-{'-'*6}")

for p in products:
    print(f"{p['id']:<4} | {p['title']:<30} | {p['category']:<14} | ${p['price']:<8} | {p['rating']}")


# ─────────────────────────────────────────
# STEP 2 — Filter (Rating ≥ 4.5) & Sort by Price (Descending)
# ─────────────────────────────────────────

print("" + "=" * 70)
print("STEP 2 — Products with Rating ≥ 4.5 (Sorted by Price: High → Low)")
print("=" * 70)

# Filter products with rating >= 4.5
filtered = [p for p in products if p["rating"] >= 4.5]

# Sort by price in descending order
sorted_products = sorted(filtered, key=lambda p: p["price"], reverse=True)

print(f"{'ID':<4} | {'Title':<30} | {'Category':<14} | {'Price':<9} | {'Rating'}")
print(f"{'-'*4}-+-{'-'*30}-+-{'-'*14}-+-{'-'*9}-+-{'-'*6}")

for p in sorted_products:
    print(f"{p['id']:<4} | {p['title']:<30} | {p['category']:<14} | ${p['price']:<8} | {p['rating']}")

print(f"Total products with rating ≥ 4.5: {len(sorted_products)}")


# ─────────────────────────────────────────
# STEP 3 — Search by Category: Laptops
# ─────────────────────────────────────────

print("" + "=" * 70)
print("STEP 3 — Fetching All Products in 'laptops' Category")
print("=" * 70)

laptop_url = "https://dummyjson.com/products/category/laptops"
laptop_response = requests.get(laptop_url)
laptop_data = laptop_response.json()

laptops = laptop_data["products"]

print(f"Found {len(laptops)} laptop(s):")
print(f"{'#':<4} | {'Name':<35} | {'Price'}")
print(f"{'-'*4}-+-{'-'*35}-+-{'-'*10}")

for i, laptop in enumerate(laptops, start=1):
    print(f"{i:<4} | {laptop['title']:<35} | ${laptop['price']}")


# ─────────────────────────────────────────
# STEP 4 — POST Request (Simulated Add Product)
# ─────────────────────────────────────────

print("" + "=" * 70)
print("STEP 4 — Sending POST Request to Add a Custom Product")
print("=" * 70)

post_url = "https://dummyjson.com/products/add"

new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

post_response = requests.post(post_url, json=new_product)
result = post_response.json()

print("Server Response:")
print("-" * 40)
for key, value in result.items():
    print(f"  {key}: {value}")

print("Note: DummyJSON is a mock API. The product is NOT actually stored.")
print("      The server simulates the response and assigns a fake ID.")
print("All 4 steps completed successfully!")
```

```
# ============================================================
#   TASK 3 — Exception Handling
#   Theme: Product Explorer & Error-Resilient Logger
# ============================================================

import requests

# ─────────────────────────────────────────
# PART A — Guarded Calculator: safe_divide()
# ─────────────────────────────────────────

print("=" * 60)
print("PART A — Guarded Calculator")
print("=" * 60)

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))


# ─────────────────────────────────────────
# PART B — Guarded File Reader
# ─────────────────────────────────────────

print("=" * 60)
print("PART B — Guarded File Reader")
print("=" * 60)

def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as file:
            content = file.read()
        return content
    except FileNotFoundError:
        print("Error: File '" + filename + "' not found.")
        return None
    finally:
        print("File operation attempt complete.")

print("--- Reading python_notes.txt ---")
content = read_file_safe("python_notes.txt")
if content:
    print(content)

print("--- Reading ghost_file.txt ---")
read_file_safe("ghost_file.txt")


# ─────────────────────────────────────────
# PART C — Robust API Calls
# ─────────────────────────────────────────

print("=" * 60)
print("PART C — Robust API Calls")
print("=" * 60)

def safe_get(url, timeout=5):
    try:
        response = requests.get(url, timeout=timeout)
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
        return None
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
        return None
    except Exception as e:
        print("Unexpected error: " + str(e))
        return None

def safe_post(url, payload, timeout=5):
    try:
        response = requests.post(url, json=payload, timeout=timeout)
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
        return None
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
        return None
    except Exception as e:
        print("Unexpected error: " + str(e))
        return None

print("Fetching 20 products...")
response = safe_get("https://dummyjson.com/products?limit=20")
if response:
    products = response.json()["products"]
    print("Fetched " + str(len(products)) + " products successfully.")
    filtered = sorted(
        [p for p in products if p["rating"] >= 4.5],
        key=lambda p: p["price"],
        reverse=True
    )
    print("Products with rating >= 4.5: " + str(len(filtered)))

print("Fetching laptops category...")
laptop_response = safe_get("https://dummyjson.com/products/category/laptops")
if laptop_response:
    laptops = laptop_response.json()["products"]
    print("Found " + str(len(laptops)) + " laptop(s).")

print("Sending POST request...")
new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}
post_response = safe_post("https://dummyjson.com/products/add", new_product)
if post_response:
    result = post_response.json()
    print("Server Response:")
    for key, value in result.items():
        print("  " + str(key) + ": " + str(value))


# ─────────────────────────────────────────
# PART D — Input Validation Loop
# ─────────────────────────────────────────

print("=" * 60)
print("PART D — Product Lookup by ID (1-100)")
print("=" * 60)

while True:
    print("")
    user_input = input("Enter a product ID to look up (1-100), or 'quit' to exit: ").strip()

    if user_input.lower() == "quit":
        print("Goodbye! Exiting product lookup.")
        break

    try:
        product_id = int(user_input)
    except ValueError:
        print("Warning: Please enter a valid whole number, not text.")
        continue

    if product_id < 1 or product_id > 100:
        print("Warning: ID must be between 1 and 100. Please try again.")
        continue

    lookup_url = "https://dummyjson.com/products/" + str(product_id)
    response = safe_get(lookup_url)

    if response is None:
        continue

    if response.status_code == 404:
        print("Product not found.")
    elif response.status_code == 200:
        product = response.json()
        print("  Title : " + product["title"])
        print("  Price : $" + str(product["price"]))
    else:
        print("  Unexpected response code: " + str(response.status_code))
```

```
# ============================================================
#   TASK 4 — Logging to File
#   Theme: Product Explorer & Error-Resilient Logger
# ============================================================

import requests
from datetime import datetime

# ─────────────────────────────────────────
# HELPER — Write a timestamped log entry
# ─────────────────────────────────────────

def log_error(function_name, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log_entry = "[" + timestamp + "] ERROR in " + function_name + ": " + error_type + " — " + message
    with open("error_log.txt", "a", encoding="utf-8") as log_file:
        log_file.write(log_entry + "")
    print("Logged: " + log_entry)


# ─────────────────────────────────────────
# TEST 1 — Trigger a ConnectionError
# Using a genuinely unreachable URL
# ─────────────────────────────────────────

print("=" * 60)
print("TEST 1 — Triggering a ConnectionError")
print("=" * 60)

try:
    response = requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except requests.exceptions.ConnectionError as e:
    log_error("fetch_products", "ConnectionError", "No connection could be made")
except requests.exceptions.Timeout:
    log_error("fetch_products", "Timeout", "Request timed out after 5 seconds")
except Exception as e:
    log_error("fetch_products", "UnexpectedError", str(e))


# ─────────────────────────────────────────
# TEST 2 — Trigger an HTTP 404 Error
# Requesting a product ID that does not exist
# ─────────────────────────────────────────

print("")
print("=" * 60)
print("TEST 2 — Triggering an HTTP 404 Error (Product ID 999)")
print("=" * 60)

try:
    response = requests.get("https://dummyjson.com/products/999", timeout=5)
    if response.status_code == 404:
        log_error("lookup_product", "HTTPError", "404 Not Found for product ID 999")
    elif response.status_code == 200:
        product = response.json()
        print("Product found: " + product["title"])
    else:
        log_error("lookup_product", "HTTPError", str(response.status_code) + " Unexpected status for product ID 999")
except requests.exceptions.ConnectionError:
    log_error("lookup_product", "ConnectionError", "No connection could be made")
except requests.exceptions.Timeout:
    log_error("lookup_product", "Timeout", "Request timed out after 5 seconds")
except Exception as e:
    log_error("lookup_product", "UnexpectedError", str(e))


# ─────────────────────────────────────────
# READ AND PRINT THE FULL LOG FILE
# ─────────────────────────────────────────

print("")
print("=" * 60)
print("FULL CONTENTS OF error_log.txt")
print("=" * 60)

try:
    with open("error_log.txt", "r", encoding="utf-8") as log_file:
        contents = log_file.read()
    print(contents)
except FileNotFoundError:
    print("No log file found yet.")
```
