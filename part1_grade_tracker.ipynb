```python
# Student Grade Tracker - Task 1

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

print("TASK 1: CLEANING STUDENT DATA")
print("=" * 50)
print()

# list to store cleaned student data
cleaned_students = []

print("Starting to clean data for each student...")
print()

for student in raw_students: 
    # loop through each student in raw data
    print(f"Processing student: {student['name']}")
    
    # Clean the name: remove extra spaces and capitalize properly
    name = student["name"].strip().title()  
    print(f"  > Cleaned name: {name}")
    
    # convert roll number from string to integer
    roll = int(student["roll"])
    print(f"  > Roll number: {roll}")
    
    # convert marks from string to list of integers
    marks = []
    for m in student["marks_str"].split(", "):
        marks.append(int(m))
    print(f"  > Marks converted: {marks}")

    # check if name contains only alphabetic characters
    valid = True
    for word in name.split():
        if not word.isalpha():
            valid = False

    # print validation result
    if valid:
        print(f"  [OK] Valid name: {name}")
    else:
        print(f"  [X] Invalid name: {name}")
    
    # store cleaned data in dictionary and add to list
    cleaned_students.append({"name": name, "roll": roll, "marks": marks})
    
    # print formatted profile card for each student
    print("================================")
    print(f"Student : {name}")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("================================")
    print()

print("Now searching for student with Roll No 103...")
print()

# find student with roll number 103 and print name in different cases
for student in cleaned_students:
    if student["roll"] == 103:
        print(f"Found! Student with Roll No 103:")
        print(f"ALL CAPS: {student['name'].upper()}")
        print(f"lowercase: {student['name'].lower()}")

print()
print()

```

```python
# Task 2 - Marks Analysis Using Loops & Conditionals
print("TASK 2: MARKS ANALYSIS FOR INDIVIDUAL STUDENT")
print("=" * 50)
print()

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

print(f"Analyzing marks for: {student_name}")
print("=" * 50)
print()

print("Calculating grade for each subject...")
print()

# Part 1: Print each subject with marks and grade label
for i in range(len(subjects)):
    subject = subjects[i]
    mark = marks[i]

    # determine grade based on marks
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"
    
    print(f"{subject}: {mark} - Grade {grade}")

print("=" * 50)
print()
print()

```

```python
# Task 3 - Class Performance Summary
print("TASK 3: CLASS PERFORMANCE SUMMARY")
print("=" * 50)
print()

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print("Calculating average marks and status for all students...")
print()

# print table header
print("Name              |Average | Status")
print("-" * 40)

# tracking class statistics
pass_count = 0
fail_count = 0
topper_name = ""
topper_avg = 0
all_averages = []

# loop through each student data
for student in class_data:
    name = student[0]
    marks = student[1]

    # calculate average marks
    total = 0
    for mark in marks:
        total += mark
    average = round(total / len(marks), 2)

    # store average for class average calculation
    all_averages.append(average)

    # Evaluate if pass or fail
    if average >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    # check if this is the class topper
    if average > topper_avg:
        topper_avg = average
        topper_name = name

    # print student row with proper spacing
    print(f"{name:18}| {average:7.2f} | {status}")

# calculate class average
class_total = 0
for avg in all_averages:
    class_total += avg
class_average = round(class_total / len(all_averages), 2)

# print class report
print()
print("CLASS SUMMARY:")
print(f"  > Students who passed: {pass_count}")
print(f"  > Students who failed: {fail_count}")
print(f"  > Class topper: {topper_name} ({topper_avg})")
print(f"  > Class average: {class_average}")
print()
print()

```

```python
# Task 4-String Manipulation Utility 
print("TASK 4: STRING MANIPULATION UTILITY")
print("=" * 50)
print()

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

print("Original essay (with extra spaces):")
print(f"'{essay}'")
print()

# Step 1: Strip whitespace and store as clean_essay
clean_essay = essay.strip()
print("Step 1 - Removing extra spaces from beginning and end:")
print(f"'{clean_essay}'")
print()

# Step 2: Convert to Title Case
title_case_essay = clean_essay.title()
print("Step 2 - Converting to Title Case:")
print(title_case_essay)
print()

# Step 3: Count how many times "python" appears
python_count = clean_essay.count("python")
print("Step 3 - Counting occurrences of 'python':")
print(f"The word 'python' appears {python_count} times")
print()

# Step 4: Replace "python" with "Python [snake emoji]"
replaced_essay = clean_essay.replace("python", "Python [snake]")
print("Step 4 - Replacing 'python' with 'Python [snake]':")
print(replaced_essay)
print()

# Step 5: Split into sentences
sentences = clean_essay.split(". ")
print("Step 5 - Splitting essay into separate sentences:")
print(sentences)
print()

# Step 6: Print each sentence numbered with period at end
print("Step 6 - Printing numbered sentences:")
for i in range(len(sentences)):
    sentence = sentences[i]
    # check if sentence already ends with period
    if not sentence.endswith("."):
        sentence = sentence + "."
    print(f"{i + 1}. {sentence}")

print()
print("=" * 50)
print("ALL TASKS COMPLETED!")
print("=" * 50)
```
